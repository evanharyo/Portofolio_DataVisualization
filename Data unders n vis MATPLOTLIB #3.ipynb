## ScatterPlots

```python
plt.figure(figsize=(5,4),dpi=100)
x = np.random.randn(100)
y = np.random.randn(100)
colors = np.random.rand(100)
sizes = 100*np.random.rand(100)
plt.scatter(x,y,c=colors,s=sizes,alpha=0.3,cmap=plt.cm.magma)
plt.colorbar()
```

## Histogram

```python
plt.figure(figsize=(4,3))
d1 = np.random.normal(0,0.7,1000)
d2 = np.random.normal(-5,2,1000)
d3 = np.random.normal(5,3,1000)
plt.hist(d1)
plt.hist(d2)
plt.hist(d3)
```

## 3D using Matplotlib

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure()
ax = fig.gca(projection='3d')
theta = np.linspace(-4*np.pi,4*np.pi,100)
Z = np.linspace(-2,2,100)
r = Z**2+1
X = r*np.sin(theta)
Y= r*np.cos(theta)
ax.plot(X,Y,Z,label='3d')
ax.legend()
ax.set_title('3d')
ax.set_xlabel('x')
ax.set_ylabel('y')
```
