```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

#difference between seaborn n matplotlib-> https://analyticsindiamag.com/comparing-python-data-visualization-tools-matplotlib-vs-seaborn/
```

```python
df = pd.DataFrame({'x':np.random.normal(0,2,1000),
                  'y': np.random.normal(3,5,1000),
                  'z': np.random.normal(-4,10,1000)})
```

```python
for col in df.columns:
    plt.hist(df[col],density=True, alpha=0.5)
    sns.kdeplot(df[col],shade=True)
```

```python
sns.set
```

```python
for col in df.columns:
    sns.distplot(df[col])
```

```python
with sns.axes_style('white'):
    sns.jointplot('x','y',df,kind='kde')
```

```python
iris = sns.load_dataset('iris')
sns.pairplot(iris,hue='species')
```

## Seaborn Relplot

```python
iris.head()
```

```python
sns.relplot(x='sepal_length',y='sepal_width',data=iris) #mirip scatter plot
```

```python
sns.relplot(x='sepal_length',y='sepal_width',data=iris, hue='species')
```

```python
sns.relplot(x='sepal_length',y='sepal_width',data=iris, hue='species',
           style='species')
```

## Seaborn Catplot
